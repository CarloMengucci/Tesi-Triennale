```python
from __future__ import division
import pylab as plt
import numpy as np
import scipy as sp
import scipy.stats as st
import pandas as pd
import sklearn
import patsy
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.cross_validation import train_test_split
```

```python
#Funzione che ritorna arrays contenenti indici per lo splitting in train e test (randomizzata)
def rand_tt_ar_split_idx(size,percent,train_percent,start_point=0):        
      
        tot_index=np.arange(size)
        slice_size=size*percent//100
        slice_index=np.arange(start_point,slice_size)
        train_idx_size=train_percent/100
        
        
        train_index,test_index=train_test_split(slice_index,train_size=train_idx_size)
            
        test_index_l=[]
            
        for i in test_index:
                test_index_l.append(i)
            
          
            
        for i in tot_index: 
            if i not in train_index:
                if i not in test_index_l:        
                    test_index_l.append(i)
            
        test_index_tot=np.array(test_index_l)
        test_index_l=[]
            
        return (train_index,test_index_tot)
```

```python
#Funzione di Learning avanzato per troncamento dati (compatibile con dmatrices)-->utilizzo di attributi iloc
def tt_adv_split_learning(x,y,slice_percent,train_percent,n_steps):
    
   
    inc=len(x)//n_steps
    
    it_count=0
    sp=slice_percent
    tp=train_percent
    correlators=[]
    residuals=[]
 
    for i in range (n_steps):
        
        
          
        
        for i in range (10):

            tr_index,ts_index=rand_tt_ar_split_idx(len(x),sp,tp,it_count)
            
            
            xdata_train=x.iloc[tr_index]
            xdata_test=x.iloc[ts_index]
    
            ydata_train=y.iloc[tr_index]
            ydata_test=y.iloc[ts_index]
            
            if len(xdata_train)|len(ydata_train)==0:
                break
            
            linreg=LinearRegression(fit_intercept=False)
            linreg.fit(xdata_train,ydata_train)
        
            correlatori=linreg.score(xdata_test,ydata_test)
            correlators.append(correlatori)
                    
        c=it_count+inc
        it_count=c
    
    mean=np.mean(correlators)
    std_dev=np.std(correlators)
   
    return (mean) 
```

```python
#linee di costruzione dell'ambiente di testing, l'idea è quella di valutare l'efficacia della funzione, prima con tagli
#randomici ed in seguito con tagli deterministici, ambiente nel quale dovrebbe fornire i risultati migliori



from patsy import dmatrices, dmatrix
from sklearn.cross_validation import cross_val_score
x1 = plt.randn(100)
x2 = plt.randn(100)
x3 = plt.randn(100)
y = 0.5*x1 - 2*x2 +x1*x2+x2*x3+ 0.5*plt.randn(100)
data = pd.DataFrame({'x1':x1, 'x2':x2,'x3':x3, 'y':y})

modelli = ["y ~ x1",
           "y ~ x2",
           "y ~ x1 + x2",
           "y ~ x1 * x2",
           "y~x1+x3",
           "y~x2*x3",
           "y~x1*x3",
           "y~x3"
           ]
```

```python
#Sezione di prova con l'intero database, funzione standard
scores = {}
for formula in modelli:
    outcome, predictors = dmatrices(formula, data)
   
  
    linreg = LinearRegression(fit_intercept=False)
    
    model_scores = cross_val_score(linreg, predictors, outcome, cv=10)
    scores[formula] = model_scores

scores_mean = {k:np.mean(v) for k, v in scores.items()}


results=pd.DataFrame({'scores':scores_mean.items()})
#print(scores)
#print()
print(results)
```

```python
#Sezione di prova con l'intero database, funzione avanzata

scores = {}
for formula in modelli:
    outcome, predictors = dmatrices(formula, data,return_type='dataframe')
    tt_adv_split_learning(predictors,outcome,100,80,3)
    
    
    
    model_scores = tt_adv_split_learning(predictors,outcome,100,80,3)
    
    scores[formula] = model_scores

scores_mean = {k:np.mean(v) for k, v in scores.items()}
results=pd.DataFrame({'scores':scores_mean.items()})
#print(scores)
#print()
print(results)
```
